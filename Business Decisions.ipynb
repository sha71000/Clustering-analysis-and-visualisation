```python
import pandas as pd
df=pd.read_csv('toronto.csv')
df=df.drop('Unnamed: 0',1)
df
```

# USER PERSONA:

Bob owns a chain of Breweries and Bars. He wishes to expand his chain and open a few lounges in Toronto City. He wants to study the city neighborhoods and decide on the best locations to open his businesses in.

```python
#Function to return neighborhoods that have no Bar/Pub/Brewery near them
def noBar(d):
    
    noBarHere=d[~d['1st Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['2nd Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['3rd Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['4th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['5th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['6th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['7th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['8th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['9th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['10th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])]
    return noBarHere


```

```python
noneHere=noBar(df)
noneHere
```

# Visualisation of the above table:




```python
import folium
map_noBar = folium.Map(location=[43.6534817, -79.3839347], zoom_start=11)

for lat, lng, label in zip(noneHere['latitude'], noneHere['longitude'], noneHere['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_noBar)  
    
map_noBar
```

Any of the above locations will be a good location to open up a bar/Pub since there would be minimal competitors
