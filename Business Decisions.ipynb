```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df=pd.read_csv('toronto.csv')
df=df.drop('Unnamed: 0',1)
df
```

# USER PERSONA 1:

Bob owns a chain of Breweries and Bars. He wishes to expand his chain and open a few lounges in Toronto City. He wants to study the city neighborhoods and decide on the best locations to open his businesses in.

```python
#Function to return neighborhoods that have no Bar/Pub/Brewery near them
def noBar(d):
    
    noBarHere=d[~d['1st Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['2nd Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['3rd Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['4th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['5th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['6th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['7th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['8th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['9th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])&
                ~d['10th Most Common Venue'].isin(['Brewery','Pub','Bar','Gastropub','Beer Bar'])]
    return noBarHere


```

```python
noneHere=noBar(df)
noneHere
```

# Visualisation of areas with no Bars:




```python
import folium
map_noBar = folium.Map(location=[43.6534817, -79.3839347], zoom_start=11)

for lat, lng, label in zip(noneHere['latitude'], noneHere['longitude'], noneHere['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_noBar)  
    
map_noBar
```

Any of the above locations will be a good location to open up a bar/Pub since there would be minimal competitors


```python
allVenues=pd.read_csv('allVenues.csv')
allVenues=allVenues.drop(['Unnamed: 0','Neighborhood Latitude','Neighborhood Longitude','Venue Latitude','Venue Longitude'],axis=1)
allVenues
```

# Neighborhoods by volume of Venues

```python
allVenues.groupby('Neighborhood').count().sort_values(by="Venue",ascending=False).reset_index().drop('Venue Category',axis=1)
```

# Total Venues by Category in Toronto

```python
toronto_venue_count=allVenues['Venue Category'].value_counts().reset_index()

toronto_venue_count.rename(columns={'index':'Venue Category','Venue Category':'Count'},inplace=True)
toronto_venue_count.set_index(['Venue Category'],inplace=True)
```

# USER PERSONA 2:

# Restaurant Category Visualisation 

```python
restaurants=toronto_venue_count[toronto_venue_count.index.map(lambda x: x.endswith('Restaurant'))]
```

```python
restaurants=restaurants.iloc[1:]
restaurants.drop(restaurants.index[3],inplace=True)
```

```python
restaurants=restaurants[~restaurants.index.str.contains("Asian")&
            ~restaurants.index.str.contains("Vegan")&
            ~restaurants.index.str.contains("American")&
            ~restaurants.index.str.contains("Seafood")&
            ~restaurants.index.str.contains("Falafel")&
            ~restaurants.index.str.contains("European")&
            ~restaurants.index.str.contains("Comfort")&
            ~restaurants.index.str.contains("Gluten")&
            ~restaurants.index.str.contains("Szechuan")&
            ~restaurants.index.str.contains("Tapas")&
            ~restaurants.index.str.contains("Dongbei")&
            ~restaurants.index.str.contains("Xinjiang")&
            ~restaurants.index.str.contains("Dumpli")&
            ~restaurants.index.str.contains("Dim")&
            ~restaurants.index.str.contains("Ramen")&
            ~restaurants.index.str.contains("Theme")&
            ~restaurants.index.str.contains("Hakka")&
            ~restaurants.index.str.contains("Fast")
            
           
           ]            
```

```python
import matplotlib.pyplot as plt
```

```python
r=list(restaurants.index)
labels=[item.split(" ")[0] for item in r]
```

```python
ax=plt.subplot()
restaurants.iloc[:15].plot(kind="barh",figsize=(10, 6),ax=ax)
ax.set_yticks(range(len(labels[:15])))
ax.set_yticklabels(labels[:15])
plt.xlabel('Number')
plt.ylabel('Restaurant Categories')
plt.title('Most Popular Restaurants in Toronto')
plt.show()
```

```python
ax1=plt.subplot()
restaurants.iloc[15:].plot(kind="barh",figsize=(10, 6),ax=ax1)
ax1.set_yticks(range(len(labels[15:])))
ax1.set_yticklabels(labels[15:])
plt.xlabel('Number')
plt.ylabel('Restaurant Categories')
plt.title('Least Popular Restaurants in Toronto')
plt.show()
```
